# barplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# collect data from seaborn data set (github)
df=sns.load_dataset('penguins')
df
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df)
plt.show()
```

```python
# using hue parameter for separated
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex')
plt.title('barplot')
plt.show()
```

```python
# using order parameter
order1=['Dream','Torgersen','Biscoe']
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',order=order1)
plt.title('barplot')
plt.show()
```

```python
# changing bar (female than male) through the hue parameter
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'])
plt.title('barplot')
plt.show()
```

```python
# changing the interval line through the (ci) parameter and it has range (0 to 100)
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],ci=100)
plt.title('barplot')
plt.show()
```

```python
# changing the interval line through the (ci) parameter and it has range (0 to 100)
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],ci=0)
plt.title('barplot')
plt.show()
```

```python
# changing the interval line through the (ci) parameter and it has range (0 to 100)
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],ci=33)
plt.title('barplot')
plt.show()
```

```python
# using orientation for horizontal
sns.barplot(x='island',y='bill_length_mm',data=df,orient='h')
plt.title('barplot')
plt.show()
```

```python
sns.barplot(x='island',y='bill_length_mm',data=df,color='b')
plt.show()
```

```python
# using palette parameter for color
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],palette='icefire')
plt.title('barplot')
plt.show()
```

```python
# using palette parameter for color with saturation
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],palette='icefire',
            saturation=0.2)
plt.title('barplot')
plt.show()
```

```python
# error color and error with (interval line)
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],palette='icefire',
            errcolor='k',errwidth='4')
plt.title('barplot')
plt.show()
```

```python
#capsize(remember boxplot)
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],palette='icefire',
            errcolor='b',capsize=0.5)
plt.title('barplot')
plt.show()
```

```python
# using darkgrid and alpha parameter
sns.set(style='darkgrid')
sns.barplot(x='island',y='bill_length_mm',data=df,hue='sex',hue_order=['Female','Male'],palette='icefire',
            errcolor='b',alpha=0.7)
plt.title('barplot')
plt.show()
```
