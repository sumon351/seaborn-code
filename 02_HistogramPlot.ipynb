# histogram plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
df
```

```python
# create histplot(using displot)
sns.displot(x='bill_length_mm',data=df)
plt.show()
```

```python
# create histplot(using displot)
sns.displot(x='bill_depth_mm',data=df)
plt.show()
```

```python
sns.displot(x='flipper_length_mm',data=df,kde=True)
plt.show()
```

```python
# using (bins) parameter for arranging data like this histogram plot
sns.displot(x='flipper_length_mm',data=df,bins=[170,180,190,200,210,220,230])
plt.show()
```

```python
# using (bins) parameter for arranging data like this histogram plot with kde and rug
sns.displot(x='flipper_length_mm',data=df,bins=[170,180,190,200,210,220,230],kde=True,rug=True,color='r')
plt.show()
```
