# scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
df
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df)
plt.title('scatter plot')
plt.show()
```

```python
# using hue
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex')
plt.title('scatter plot')
plt.show()
```

```python
df=sns.load_dataset('penguins').head(20)
df
```

```python
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex')
plt.title('scatter plot')
plt.show()
```

```python
# style
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',style='sex')
plt.title('scatter plot')
plt.show()
```

```python
# color parameter
sns.scatterplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',style='sex',
               palette='Accent')
plt.title('scatter plot')
plt.show()
```
